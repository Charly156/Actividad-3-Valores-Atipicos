```python
#Se importan las bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Se lee la base de datos
data = pd.read_csv('Ventas_totales_limpio.csv')

#Se elimina una columna que esta dem√°s
data= data.drop(columns = ['Unnamed: 0'])
data
```

```python
#Corroboramos valores nulos de la base de datos para poder analizar los datos
valores_nulos = data.isnull().sum()
valores_nulos
```

```python
#Se hace una copia del dataframe por seguridad
data1 = data.copy()
```

```python
#Usamos la grafica de boxplot en una de las columnas para ver su distribuci√≥n
fig = plt.figure(figsize = (10,6))
plt.boxplot(data['subtotal_ventas_alimentos_bebidas'])
plt.title('Subtotal de ventas de alimentos y bebidas')
plt.grid()
```

```python
#Viendo que hay outliers en la gr√°fica, se usa el
#m√©todo de cuartiles. Encuentro cuartiles 0.25 y 0.75
y = data1['subtotal_ventas_alimentos_bebidas']

percentile25 = y.quantile(0.25) #Q1
percentile75 = y.quantile(0.75) #Q3
iqr = percentile75 - percentile25

Limite_superior_iqr = percentile75 + 1.5 * iqr
Limite_inferior_iqr = percentile25 - 1.5 * iqr
print('Limite superior permitido', Limite_superior_iqr)
print('Limite inferior permitido', Limite_inferior_iqr)
```

```python
#Obtenemos datos limpios y los almacenamos en otra variable
data_clean_iqr = data1[(y<=Limite_superior_iqr)&(y>=Limite_inferior_iqr)]
registro1 = data_clean_iqr.copy()
data_clean_iqr
```

```python
#Se crea la caja de boxplot con los datos limpios
fig = plt.figure(figsize = (10,6))
plt.boxplot(data_clean_iqr['subtotal_ventas_alimentos_bebidas'])
plt.title('Subtotal de ventas de alimentos y bebidas')
plt.grid()
plt.show()
```

##Se repite el proceso

```python
fig = plt.figure(figsize = (10,6))
plt.boxplot(data['tarjetas_debito'])
plt.title('Tarjetas de debito')
plt.grid()
```
